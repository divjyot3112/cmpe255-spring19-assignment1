```python
positive = {'good' : True, 'great' : True, 'best' : True, 'perfect' : True, 'excellent' : True, 'amazing' : True, 'fresh' : True,
       'satisfied' : True, 'delicious' : True, 'perfectly' : True, 'like' : True, 'awesome' : True, 'loved' : True, 
       'love' : True, 'delight' : True, 'delightful' : True, 'grace' : True, 'quick' : True, 'favorite' : True,
       'tasty' : True, 'crisp' : True, 'crispy' : True, 'enjoy' : True, 'impressed' : True, 'satifaction' : True,
       'definitely' : True, 'yummy' : True }
negative = {'bad' : True, 'worse' : True, 'worst' : True, 'hate' : True, 'hated' : True, 'regret' : True, 'awful' : True,
       'stale' : True, 'soggy' : True, 'dirty' : True, 'sadly' : True, 'waited' : True, 'wait' : True, 'slow' : True,
       'mistake' : True, 'embarressed' : True, 'disappointed' : True, 'unimpressed' : True, 'disappointment' : True,
       'waiting' : True, 'nothing' : True }
neutral = {'OK' : True, 'ok' : True, 'fine' : True, 'alright' : True, 'but' : True, 'fair' : True, 'decent' : True,
       'limited' : True }
```

```python
negative = negative.keys()
positive = positive.keys()
neutral = neutral.keys()
clf = LogisticRegression()
vectorizer = HashingVectorizer(analyzer='word',lowercase=False)
```

```python
def getClassification(s):
    pcount = 0
    ncount = 0
    nucount = 0
    s = s.lower()
    for t in positive:
      if(t.lower() in s):
        pcount = pcount + 1
    for t in negative:
      if(t.lower() in s):
        ncount = ncount + 1
    for t in neutral:
      if(t.lower() in s):
        nucount = nucount + 1
    max_count = max(pcount,ncount,nucount)
    if(pcount-ncount>0):
      return 'positive'
    elif(ncount-pcount>0):
      return 'negative'
    return 'neutral'
```

```python
df2['rate'] = df2['text'].apply(getClassification)
features_nd = vectorizer.fit_transform((df2['text']))
```

```python
df2.loc[df2['rate']=='positive','rate'] = 3
df2.loc[df2['rate']=='negative','rate'] = 1
df2.loc[df2['rate']=='neutral','rate'] = 2
df2.sort_values('business_id')
y_train = df2['rate']
# df2
```

```python
X_train, X_test, y_train, y_test  = train_test_split(features_nd, np.asarray(y_train,dtype=np.float64),
                                                    train_size=0.80, random_state=1234,shuffle=False)
```

```python
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
y_true = y_test
print('accuracy  - ' + str(accuracy_score(y_true,y_pred)))
print('precision - ' + str(precision_score(y_true, y_pred, average='macro')))
print('recall    - ' + str(recall_score(y_true, y_pred, average='macro')))
print('cross_val - ' + str(cross_val_score(clf, features_nd, np.asarray(df2['rate'],dtype=np.float64), cv=4)))
```

```python
def getPrediction(s):
  rval = clf.predict(vectorizer.fit_transform([s]))
  return rval[0]
df2['pred'] = df2['text'].apply(getPrediction)
df2
```

```python
df_final = df2.groupby(['business_id','review_id','stars'])['pred'].mean()
df_final = pd.DataFrame({'business_id':df_final.index,'pred':df_final.values})
df_final
```
